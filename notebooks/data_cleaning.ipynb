```python
import pandas as pd
```

```python
import os
```

```python
# Folder path where your 4 CSVs are located
```

```python
path = r"C:\Users\Tridib Sarkar\Desktop\Nifty50_Project"
```

```python
#List of your CSV filenames
```

```python
files = ["TCS.csv", "INFY.csv", "RELIANCE.csv", "HDFCBANK.csv"]
```

```python
# Create an empty list to store data
```

```python
dataframes = []
```

```python
#Loop through and add 'Company' column
```

```python
for file in files:
    df = pd.read_csv(os.path.join(path, file))
    company_name = file.replace(".csv", "")
    df["Company"] = company_name
    dataframes.append(df)
```

```python
# Merge all DataFrames
```

```python
merged_df = pd.concat(dataframes,ignore_index=True)
```

```python
#Save the combined dataset
```

```python
merged_df.to_csv("clean_nifty50.csv", index=False)
```
